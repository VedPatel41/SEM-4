```python
import requests 

response = requests.get(r'https://subslikescript.com/movie/Batman_The_Enemy_Within-7170214')
```

```python
print(response)
print(response.status_code)
```

```python
content = response.text
print(content)
```

```python
from bs4 import BeautifulSoup
data = BeautifulSoup(content, 'html.parser')
print(data.prettify())
```

```python
title = data.find('h1')
print(title)
```

```python
script = data.find('div', class_='full-script')
print(script.text)
```

```python
response = requests.get(r'https://subslikescript.com/movies')
print(response.status_code)
```

```python
content = response.text
print(content)
```

```python
data = BeautifulSoup(content, 'html.parser')
article = data.find('article',class_='main-article')
print(article)
```

```python
l = article.find_all('a')
print(len(l))
```

```python
list = []
for  i in l:
    list.append(i['href'])
print(list)
```

```python
for i in list:
    url = 'https://subslikescript.com' + i
    response = requests.get(url)
    data = BeautifulSoup(response.text, 'html.parser')
    title = data.find('h1').text
    script = data.find('div', class_='full-script').text
    f = open(title+'.txt','w',encoding='utf-8')
    f.write(script)
    print(f'{title} file created successfully')
    f.close()
```

```python

user = int(input('page no'))
for  i in range((user)):
    url = 'https://subslikescript.com/movies?page=' + str(i+1)
    response = requests.get(url)
    content = response.text
    data = BeautifulSoup(content, 'html.parser')
    article = data.find('article',class_='main-article')
    l = article.find_all('a')
    link= []
    for  i in l:
        link.append(i['href'])
    for i in link:
        url = 'https://subslikescript.com' + i
        response = requests.get(url)
        data = BeautifulSoup(response.text, 'html.parser')
        title = data.find('h1').text
        script = data.find('div', class_='full-script').text
        f = open(title+'.txt','w',encoding='utf-8')
        f.write(script)
        print(f'{title} file created successfully')
        f.close()
    
```
